```python
# ================================
# Workplace Accidents Analysis
# Using Pandas, Matplotlib, Seaborn
# ================================
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# ================================
# Load the dataset
# ================================
# Replace with your actual CSV/Excel file
df = pd.read_csv("incidents.csv")

# Create a folder to save charts
output_folder = "charts"
os.makedirs(output_folder, exist_ok=True)
```

```python
# ================================
# 1. Total number of incidents
# ================================
total_incidents = df.shape[0]
print("Total Incidents:", total_incidents)
```

```python
# ================================
# 2. Fatal vs Nonfatal injuries
# ================================
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Degree of Injury", palette="Set2")
plt.title("Fatal vs Nonfatal Injuries")
plt.savefig(os.path.join(output_folder, "fatal_vs_nonfatal.png"))
plt.show()
```

```python
# ================================
# 3. Most common nature of injury
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="Nature of Injury", order=df["Nature of Injury"].value_counts().index, palette="viridis")
plt.title("Most Common Injury Types")
plt.savefig(os.path.join(output_folder, "injury_types.png"))
plt.show()
```

```python
# ================================
# 4. Most affected body parts
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="Part of Body", order=df["Part of Body"].value_counts().index, palette="mako")
plt.title("Most Frequently Injured Body Parts")
plt.savefig(os.path.join(output_folder, "body_parts.png"))
plt.show()
```

```python
# ================================
# 5. Event types leading to injuries
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="Event type", order=df["Event type"].value_counts().index, palette="cubehelix")
plt.title("Event Types Leading to Injuries")
plt.savefig(os.path.join(output_folder, "event_types.png"))
plt.show()
```

```python
# ================================
# 6. Environmental factors
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="Environmental Factor", order=df["Environmental Factor"].value_counts().index, palette="crest")
plt.title("Environmental Factors Associated with Incidents")
plt.savefig(os.path.join(output_folder, "environmental_factors.png"))
plt.show()
```

```python
# ================================
# 7. Human factors
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="Human Factor", order=df["Human Factor"].value_counts().index, palette="flare")
plt.title("Human Factors in Incidents")
plt.savefig(os.path.join(output_folder, "human_factors.png"))
plt.show()
```

```python
# ================================
# 8. Task assigned (Regular vs Non-regular)
# ================================
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Task Assigned", palette="pastel")
plt.title("Incidents by Task Assignment")
plt.savefig(os.path.join(output_folder, "task_assigned.png"))
plt.show()
```

```python
# ================================
# 9. Project type distribution
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, y="Project Type", order=df["Project Type"].value_counts().index, palette="coolwarm")
plt.title("Incidents by Project Type")
plt.savefig(os.path.join(output_folder, "project_type.png"))
plt.show()
```

```python
# ================================
# 10. Building stories distribution
# ================================
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Building Stories", palette="rocket")
plt.title("Incidents by Building Stories")
plt.savefig(os.path.join(output_folder, "building_stories.png"))
plt.show()
```

```python
# ================================
# 11. Hazardous substances involvement
# ================================
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="hazsub", palette="muted")
plt.title("Incidents Involving Hazardous Substances")
plt.xticks(rotation=90)
plt.savefig(os.path.join(output_folder, "haz_substances.png"))
plt.show()
```

```python
# ================================
# 12. Monthly distribution of incidents
# ================================
df["Event Date"] = pd.to_datetime(df["Event Date"], errors="coerce")
df["Month"] = df["Event Date"].dt.month

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Month", palette="Spectral")
plt.title("Monthly Distribution of Incidents")
plt.savefig(os.path.join(output_folder, "monthly_distribution.png"))
plt.show()
```

```python
# ================================
# 13. Amputation percentage
# ================================
amputation_count = df["Nature of Injury"].str.contains("Amputation", case=False, na=False).sum()
amputation_percentage = (amputation_count / total_incidents) * 100
print(f"Amputation Percentage: {amputation_percentage:.2f}%")
```

```python
# ================================
# 14. Top 10 Event Descriptions
# ================================
top_events = df["Event Description"].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_events.values, y=top_events.index, palette="plasma")
plt.title("Top 10 Event Descriptions")
plt.xlabel("Count")
plt.savefig(os.path.join(output_folder, "top_event_descriptions.png"))
plt.show()
```

```python
# ================================
# 15. Equipment malfunction related incidents
# ================================
malfunction = df[df["Human Factor"].str.contains("Malfunction", case=False, na=False)]
plt.figure(figsize=(6,4))
sns.countplot(data=malfunction, x="Human Factor", palette="ch:s=.25,rot=-.25")
plt.title("Incidents Related to Malfunction")
plt.xticks(rotation=90)
plt.savefig(os.path.join(output_folder, "malfunction.png"))
plt.show()
```
